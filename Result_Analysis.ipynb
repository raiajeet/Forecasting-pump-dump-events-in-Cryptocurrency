```python
import pandas as pd
random=pd.read_csv("/content/drive/MyDrive/Post_DC/P3/random_results.csv")
random.drop(columns=['Unnamed: 0'],inplace=True)
random['Data']='Random Under Sampling'
print("random.shape")
random
```

```python
outliers=pd.read_csv("/content/drive/MyDrive/Post_DC/P3/outliers_results.csv")
outliers.drop(columns=['Unnamed: 0'],inplace=True)
outliers['Data']='Outliers Removal'
print(outliers.shape)
outliers
```

```python
df=pd.concat([random,outliers],axis=0)
df=df.round(2)
print(df.shape)
df
```

```python
import matplotlib.pyplot as plt


df1 = df[df["Data"] == "Random Under Sampling"]
df2 = df[df['Data']=='Outliers Removal']

train_auc_rus = df1[df1["Set"] == "Train"]["AUC"].values
test_auc_rus = df1[df1["Set"] == "Test"]["AUC"].values
models_rus = df1[df1["Set"] == "Train"]["Model"].values

train_auc_out = df2[df2["Set"] == "Train"]["AUC"].values
test_auc_out = df2[df2["Set"] == "Test"]["AUC"].values
models_out = df2[df2["Set"] == "Train"]["Model"].values

plt.figure(figsize=(16, 8))

plt.plot(models_rus, train_auc_rus, marker='o', label='Train AUC (Random Under Sampling)', color='blue')
plt.plot(models_rus, test_auc_rus, marker='o', label='Test AUC (Random Under Sampling)', color='orange')

plt.plot(models_out, train_auc_out, marker='o', label='Train AUC (Outliers Removal)', color='green')
plt.plot(models_out, test_auc_out, marker='o', label='Test AUC (Outliers Removal)', color='red')

plt.title("AUC Comparison Across Models", fontsize=16)
plt.xlabel("Models", fontsize=14)
plt.ylabel("AUC", fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)

plt.legend(fontsize=12)
plt.grid(alpha=0.5)

plt.tight_layout()
plt.show()
```
